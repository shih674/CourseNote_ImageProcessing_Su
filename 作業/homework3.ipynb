```python
# 來自於電腦內部的影片檔，而不是串流
p1 = cv2.VideoCapture('homework3.mp4')
total_frame = p1.get(7)

counter = 1
while p1.isOpened():
    counter  = (counter+1) % total_frame
    ret, m1 = p1.read()
    if ret:
        image_gray = cv2.cvtColor(m1,cv2.COLOR_BGR2GRAY)
        result = cv2.adaptiveThreshold(image_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 21)
        cont0, herrican0 = cv2.findContours(result, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        #x, y, width, heigh = cv2.boundingRect(cont0[0])
        #cv2.rectangle(m1, (x,y), (width,heigh),(255,255,255),2)
        
        
        cv2.imshow('result', result)
        cv2.imshow('Image1', m1)
        if cv2.waitKey(10) != -1:
            break
    else:
        break
    print(f'\r {int(counter*100/total_frame)}%  {len(cont0)}',end= '')
cv2.destroyAllWindows()
```
